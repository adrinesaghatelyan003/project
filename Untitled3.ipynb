```python
def make_layer(m, n):
    """
    this function makes layer of neurons with given length
    
    Args:
        m    integer: The length of previous layer
        n    integer: The length of layer
        
    Return:
        arr    (ndarray(n,)):  array of dicts
    """
    
    
    arr = np.empty()
    
    for i in range(n):
        w_arr=np.empty()
        for j in range(m):
            w = np.random()
            w_arr.append(w)
            
        b = np.random()
        a = dict({'weight':w_arr, 'bias':b})
        arr.append(a);
    
    return arr
```

```python
def layer_connect(layer_in, layer, activation):
    """
    this function creates connection between 2 layers with certain activation function
    
    Args:
        layer_in    (ndarray(n,)): layer_in
        layer       (ndarray(m,)): array of dicts 
        activation    function: activation function name
        
    Return:
        out    (ndarray)
    """
    
    m = len(layer)
    out = np.zeros(m)
    
    for i in range(m):
        a = np.dot(layer_in, layer[i]['weight']) + layer[i]['bias']
        out[i] = activation(a)

    return(out)
```

```python
def cost_of_last_layer(a_in, layer, target):
    """
    this function counts the derivative of function for selected argument 
    Args:
        a_in_matrix :    (ndarray)
        layer       :    (ndarray)
        target      :    (ndarray) the actual values
        activation  :    activation function name
    Return:
        total_cost      
    """
    
    m = np.shape(a_in)
    cost = 0
    a = np.dot(layer[0]['weight'], a_in) + layer[0]['bias']
    g = (a)
    
    for i in range (m):
        costt = ( - target[i] * np.log(g)) - (1 - target[i]) * np.log(1 - g)
        cost = cost + costt
    
    total_cost = cost / m 
    
    return total_cost

    
```
